```python
match = match.merge(stadium, left_on = 'STADIUM', right_on = 'NAME')
```

```python
match
```

```python
winner=[]
for home_club,away_club, goal_home, goal_away in zip(match['HOME_TEAM'],match['AWAY_TEAM'],match['HOME_TEAM_SCORE'],match['AWAY_TEAM_SCORE']):
    if goal_home>goal_away:
        winner.append(home_club)
    elif goal_home<goal_away:
        winner.append(away_club)
    else:
        winner.append('DRAW')
```

```python
match['winner'] =winner 
```

```python
match
```

```python
match['winner'].value_counts().nlargest(10)[1:]
#teams with most wins
```

```python
match['winner'].value_counts().nlargest(10)[1:].plot(kind='bar',figsize=(15,10),color=['red','blue','yellow','green','orange','black','lime','pink','brown'])
plt.ylabel('NUMBER OF WINS',fontsize=20)
plt.xlabel('TEAMS',fontsize=20)
```
